```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
epoch = np.arange(0,3)
train_loss = np.load("/workdir/huichi/NF-C-VAE/data_save/std_NoF_train_loss.npy", allow_pickle=True)
val_loss = np.load("/workdir/huichi/NF-C-VAE/data_save/std_NoF_val_loss.npy", allow_pickle=True)
plt.plot(epoch, train_loss, '-', label='total_loss_train')
plt.plot(epoch, val_loss, '-', label='total_loss_val')
plt.yscale('log')
plt.legend()
```

```python
epoch = np.arange(0,250)
train_kl = np.load("/workdir/huichi/NF-C-VAE/data_save/6var_MAF_train_kl.npy", allow_pickle=True)
val_kl = np.load("/workdir/huichi/NF-C-VAE/data_save/6var_MAF_val_kl.npy", allow_pickle=True)
plt.plot(epoch, train_kl, '-', label='kl_loss_train')
plt.plot(epoch, val_kl, '-', label='kl_loss_val')
plt.yscale('log')
plt.legend()
```

```python
epoch = np.arange(0,1320)
plt.plot(epoch, train_loss - train_kl, '-', label='mse_loss_train')
plt.plot(epoch, val_loss - val_kl, '-', label='mse_loss_val')
plt.yscale('log')
plt.legend()
```
